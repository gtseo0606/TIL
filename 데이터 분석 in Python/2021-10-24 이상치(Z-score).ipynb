```python
import pandas as pd 

df = pd.read_csv("./Womens Clothing E-Commerce Reviews.csv")
df = df.drop(columns = ['Review Text', 'Title'])
df
```

```python
# 이상치 - Z-score
import numpy as np

def outliers(data):
    threshold = 3
    mean = np.mean(data)
    std = np.std(data)
    zscore = [(y - mean)/std for y in data]
    return np.where(np.abs(zscore)>threshold)

sub_index = outliers(df['Positive Feedback Count'])[0]
df.loc[sub_index, 'Positive Feedback Count']
```

```python
main_index = []
for i in df.index:
    if i not in sub_index:
        main_index.append(i)

df = df.loc[main_index]
df = df.reset_index(drop=True)
df
```
