```python
import pandas as pd 

df = pd.read_csv("./Womens Clothing E-Commerce Reviews.csv")
df = df.drop(columns = ['Review Text', 'Title'])
df
```

```python
df.groupby('Rating')['Age'].agg(**{'mean_age':'mean'})
```

```python
df.groupby('Rating')['Age'].agg(**{'mode_Age':lambda x:x.mode()})
```

```python
df.groupby('Rating').agg(**{'min_age':pd.NamedAgg(column='Age', aggfunc='min'),
                           'max_age':pd.NamedAgg(column='Age', aggfunc='max'),
                           'mean_positive_feedback':pd.NamedAgg(column='Positive Feedback Count', aggfunc='mean'),
                           'median_positive_feedback':pd.NamedAgg(column='Positive Feedback Count', aggfunc='median'),
                           })
```
